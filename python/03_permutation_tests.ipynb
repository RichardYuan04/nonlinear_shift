```python
from statsmodels.stats.multitest import multipletests
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
face_full = pd.read_csv('data/GMM_Face_2bk_full_results.csv')
body_full = pd.read_csv('data/GMM_Body_2bk_full_results.csv')
tool_full = pd.read_csv('data/GMM_Tool_2bk_full_results.csv')
place_full = pd.read_csv('data/GMM_Place_2bk_full_results.csv')
```

`Compare GMM: face vs. place`

```python
# Set a random seed for reproducibility
np.random.seed(42)

# Extract the ranking array for each group
face_ranks = face_full['final.rank'].to_numpy()
place_ranks = place_full['final.rank'].to_numpy()

# Observed mean difference (Face - Place)
obs_diff = np.mean(face_ranks) - np.mean(place_ranks)

# Pool all ranking values
combined_ranks = np.concatenate([face_ranks, place_ranks])
n_face = len(face_ranks)
n_place = len(place_ranks)

# Number of permutations
n_permutations = 10000
permuted_diffs = np.zeros(n_permutations)

# Permutation loop
for i in range(n_permutations):
    shuffled = np.random.permutation(combined_ranks)
    perm_face = shuffled[:n_face]
    perm_place = shuffled[n_face:]
    permuted_diffs[i] = np.mean(perm_face) - np.mean(perm_place)

# Compute the two-sided p-value
p_value = np.mean(np.abs(permuted_diffs) >= np.abs(obs_diff))

# Print the result
print(f"Observed mean difference (Face - Place): {obs_diff:.3f}")
print(f"Permutation-based p-value: {p_value:.4f}")
```

```python
# Set the plot style
sns.set(style="whitegrid", font_scale=1.5)

# Compute the two-sided p=0.01 cutoffs
lower_bound = np.percentile(permuted_diffs, 0.5)   # 0.5% left tail
upper_bound = np.percentile(permuted_diffs, 99.5)  # 0.5% right tail

# Create the figure and set resolution
plt.figure(figsize=(12, 8), dpi=300)

# Plot the histogram of the permutation distribution
sns.histplot(permuted_diffs, bins=50, kde=True, color='skyblue', edgecolor='white', alpha=0.8)

# Mark the observed difference with a red line
plt.axvline(obs_diff, color='red', linestyle='-', linewidth=2, label=f'Observed diff = {obs_diff:.3f}')

# Mark the two-sided p=0.01 cutoffs with grey lines
plt.axvline(lower_bound, color='gray', linestyle='--', linewidth=1.5, label=f'Lower cutoff = {lower_bound:.3f}')
plt.axvline(upper_bound, color='gray', linestyle='--', linewidth=1.5, label=f'Upper cutoff = {upper_bound:.3f}')

# Legend and labels
plt.title('Permutation Distribution of Mean Rank Differences\n(Face - Place)', fontsize=18)
plt.xlabel('Mean Difference under Null (Permutation)', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.legend(fontsize=14)

# Add grid lines
plt.grid(alpha=0.5)

# Show the figure
plt.tight_layout()
plt.show()
```

```python
def permutation_test_with_fdr(dataframes, dataframe_names, column='final.rank', n_permutations=10000):
    results = []
    
    # Perform pairwise permutation tests
    for i, df1 in enumerate(dataframes):
        for j, df2 in enumerate(dataframes):
            if i >= j:
                continue
            
            # Extract the column values
            values1 = df1[column].dropna().values
            values2 = df2[column].dropna().values
            
            # Observed mean difference
            obs_diff = np.mean(values1) - np.mean(values2)
            
            # Combine the data
            combined = np.concatenate([values1, values2])
            
            # Permutation test
            perm_diffs = np.zeros(n_permutations)
            for k in range(n_permutations):
                np.random.shuffle(combined)
                perm_diffs[k] = np.mean(combined[:len(values1)]) - np.mean(combined[len(values1):])
            
            # Calculate p-value
            p_value = np.mean(np.abs(perm_diffs) >= np.abs(obs_diff))
            
            # Store results
            results.append({'Group1': i, 'Group2': j, 'MeanDiff': obs_diff, 'PValue': p_value})
    
    # Convert results to DataFrame
    results_df = pd.DataFrame(results)
    
    # FDR correction
    results_df['FDR_PValue'] = multipletests(results_df['PValue'], method='fdr_bh')[1]
    
    # Create a matrix for visualization
    n = len(dataframes)
    mean_diff_matrix = np.zeros((n, n))
    p_value_matrix = np.zeros((n, n))
    
    for _, row in results_df.iterrows():
        mean_diff_matrix[int(row['Group1']), int(row['Group2'])] = row['MeanDiff']
        p_value_matrix[int(row['Group1']), int(row['Group2'])] = row['FDR_PValue']
    
    # Plot the results
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))
    sns.heatmap(mean_diff_matrix, annot=True, fmt='.2f', ax=ax[0], cmap='coolwarm', xticklabels=dataframe_names, yticklabels=dataframe_names)
    ax[0].set_title('Mean Difference')
    sns.heatmap(p_value_matrix, annot=True, fmt='.3f', ax=ax[1], cmap='viridis', xticklabels=dataframe_names, yticklabels=dataframe_names)
    ax[1].set_title('FDR-Corrected P-Values')
    plt.show()
    
    return results_df

```

```python
# for GMM_full
dataframes = [face_full, body_full, tool_full, place_full]
dataframe_names = ['face_full', 'body_full', 'tool_full', 'place_full']
results = permutation_test_with_fdr(dataframes, dataframe_names)
print(results)
```

## Compare the counts of concave vs. convex regions

```python
def permutation_test_concave_with_fdr(dataframes, dataframe_names, column='Second_Mean', n_permutations=10000):
    results = []

    # Perform pairwise permutation tests for concave regions
    for i, df1 in enumerate(dataframes):
        for j, df2 in enumerate(dataframes):
            if i >= j:
                continue

            # Count concave regions (Second_Mean < 0)
            count1 = (df1[column] < 0).sum()
            count2 = (df2[column] < 0).sum()

            # Observed difference in counts
            obs_diff = count1 - count2

            # Combine the data
            combined = np.concatenate([df1[column], df2[column]])

            # Permutation test
            perm_diffs = np.zeros(n_permutations)
            for k in range(n_permutations):
                np.random.shuffle(combined)
                perm_count1 = (combined[:len(df1)] < 0).sum()
                perm_count2 = (combined[len(df1):] < 0).sum()
                perm_diffs[k] = perm_count1 - perm_count2

            # Calculate p-value
            p_value = np.mean(np.abs(perm_diffs) >= np.abs(obs_diff))

            # Store results
            results.append({'Group1': i, 'Group2': j, 'ObsDiff': obs_diff, 'PValue': p_value})

    # Convert results to DataFrame
    results_df = pd.DataFrame(results)

    # FDR correction
    results_df['FDR_PValue'] = multipletests(results_df['PValue'], method='fdr_bh')[1]

    # Create matrices for visualization
    n = len(dataframes)
    obs_diff_matrix = np.zeros((n, n))
    fdr_pvalue_matrix = np.zeros((n, n))

    for _, row in results_df.iterrows():
        obs_diff_matrix[int(row['Group1']), int(row['Group2'])] = row['ObsDiff']
        fdr_pvalue_matrix[int(row['Group1']), int(row['Group2'])] = row['FDR_PValue']

    # Plot the results
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))
    sns.heatmap(obs_diff_matrix, annot=True, fmt='.2f', ax=ax[0], cmap='coolwarm', xticklabels=dataframe_names, yticklabels=dataframe_names)
    ax[0].set_title('Observed Difference (Concave)')
    sns.heatmap(fdr_pvalue_matrix, annot=True, fmt='.3f', ax=ax[1], cmap='viridis', xticklabels=dataframe_names, yticklabels=dataframe_names)
    ax[1].set_title('FDR-Corrected P-Values (Concave)')
    plt.show()

    return results_df

def permutation_test_convex_with_fdr(dataframes, dataframe_names, column='Second_Mean', n_permutations=10000):
    results = []

    # Perform pairwise permutation tests for convex regions
    for i, df1 in enumerate(dataframes):
        for j, df2 in enumerate(dataframes):
            if i >= j:
                continue

            # Count convex regions (Second_Mean > 0)
            count1 = (df1[column] > 0).sum()
            count2 = (df2[column] > 0).sum()

            # Observed difference in counts
            obs_diff = count1 - count2

            # Combine the data
            combined = np.concatenate([df1[column], df2[column]])

            # Permutation test
            perm_diffs = np.zeros(n_permutations)
            for k in range(n_permutations):
                np.random.shuffle(combined)
                perm_count1 = (combined[:len(df1)] > 0).sum()
                perm_count2 = (combined[len(df1):] > 0).sum()
                perm_diffs[k] = perm_count1 - perm_count2

            # Calculate p-value
            p_value = np.mean(np.abs(perm_diffs) >= np.abs(obs_diff))

            # Store results
            results.append({'Group1': i, 'Group2': j, 'ObsDiff': obs_diff, 'PValue': p_value})

    # Convert results to DataFrame
    results_df = pd.DataFrame(results)

    # FDR correction
    results_df['FDR_PValue'] = multipletests(results_df['PValue'], method='fdr_bh')[1]

    # Create matrices for visualization
    n = len(dataframes)
    obs_diff_matrix = np.zeros((n, n))
    fdr_pvalue_matrix = np.zeros((n, n))

    for _, row in results_df.iterrows():
        obs_diff_matrix[int(row['Group1']), int(row['Group2'])] = row['ObsDiff']
        fdr_pvalue_matrix[int(row['Group1']), int(row['Group2'])] = row['FDR_PValue']

    # Plot the results
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))
    sns.heatmap(obs_diff_matrix, annot=True, fmt='.2f', ax=ax[0], cmap='coolwarm', xticklabels=dataframe_names, yticklabels=dataframe_names)
    ax[0].set_title('Observed Difference (Convex)')
    sns.heatmap(fdr_pvalue_matrix, annot=True, fmt='.3f', ax=ax[1], cmap='viridis', xticklabels=dataframe_names, yticklabels=dataframe_names)
    ax[1].set_title('FDR-Corrected P-Values (Convex)')
    plt.show()

    return results_df
```

`Permutation test on concave/convex counts in face vs. place, using the GMM non-linearity split`

```python
# Define permutation test for concave regions
np.random.seed(42)

# Filter concave regions based on new definition
face_concave = face_gmm_gam_sig[(face_gmm_gam_sig['gmm_nonlinear'] == 1) & (face_gmm_gam_sig['Second_Mean'] < 0)]
place_concave = place_gmm_gam_sig[(place_gmm_gam_sig['gmm_nonlinear'] == 1) & (place_gmm_gam_sig['Second_Mean'] < 0)]

# Observed difference in concave counts
obs_diff_concave = len(face_concave) - len(place_concave)

# Combine data for permutation
data_concave = np.concatenate([
    np.ones(len(face_concave)),  # Label 1 for face
    np.zeros(len(place_concave)) # Label 0 for place
])

# Permutation test for concave
n_permutations = 10000
perm_diffs_concave = np.zeros(n_permutations)
for i in range(n_permutations):
    np.random.shuffle(data_concave)
    perm_face_count = np.sum(data_concave[:len(face_concave)])
    perm_place_count = np.sum(data_concave[len(face_concave):])
    perm_diffs_concave[i] = perm_face_count - perm_place_count

# Calculate p-value for concave
p_value_concave = np.mean(np.abs(perm_diffs_concave) >= np.abs(obs_diff_concave))

print(f"Observed difference in concave counts: {obs_diff_concave}")
print(f"Permutation-based p-value for concave: {p_value_concave:.4f}")

# Define permutation test for convex regions
# Filter convex regions based on new definition
face_convex = face_gmm_gam_sig[(face_gmm_gam_sig['gmm_nonlinear'] == 1) & (face_gmm_gam_sig['Second_Mean'] > 0)]
place_convex = place_gmm_gam_sig[(place_gmm_gam_sig['gmm_nonlinear'] == 1) & (place_gmm_gam_sig['Second_Mean'] > 0)]

# Observed difference in convex counts
obs_diff_convex = len(face_convex) - len(place_convex)

# Combine data for permutation
data_convex = np.concatenate([
    np.ones(len(face_convex)),  # Label 1 for face
    np.zeros(len(place_convex)) # Label 0 for place
])

# Permutation test for convex
perm_diffs_convex = np.zeros(n_permutations)
for i in range(n_permutations):
    np.random.shuffle(data_convex)
    perm_face_count = np.sum(data_convex[:len(face_convex)])
    perm_place_count = np.sum(data_convex[len(face_convex):])
    perm_diffs_convex[i] = perm_face_count - perm_place_count

# Calculate p-value for convex
p_value_convex = np.mean(np.abs(perm_diffs_convex) >= np.abs(obs_diff_convex))

print(f"Observed difference in convex counts: {obs_diff_convex}")
print(f"Permutation-based p-value for convex: {p_value_convex:.4f}")
```
