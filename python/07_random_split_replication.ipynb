```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
SA_rank = pd.read_csv('data/SA_rank.csv')
```

```python
face1 = pd.read_csv('data/face_subsample1_derivatives_results_label.csv')
face2 = pd.read_csv('data/face_subsample2_derivatives_results_label.csv')
place1 = pd.read_csv('data/place_subsample1_derivatives_results_label.csv')
place2 = pd.read_csv('data/place_subsample2_derivatives_results_label.csv')
```

```python
body1 = pd.read_csv('data/body_subsample1_derivatives_results_label.csv')
body2 = pd.read_csv('data/body_subsample2_derivatives_results_label.csv')
tool1 = pd.read_csv('data/tool_subsample1_derivatives_results_label.csv')
tool2 = pd.read_csv('data/tool_subsample2_derivatives_results_label.csv')
```

```python
# Extract the region name from the 'label' column
face1['region'] = face1['label'].str.split('_').str[-1]
face2['region'] = face2['label'].str.split('_').str[-1]
place1['region'] = place1['label'].str.split('_').str[-1]
place2['region'] = place2['label'].str.split('_').str[-1]

# Map the 'final.rank' from SA_rank to face1, face2, place1, and place2
region_to_rank = SA_rank.set_index('region')['final.rank']
face1['final.rank'] = face1['region'].map(region_to_rank)
face2['final.rank'] = face2['region'].map(region_to_rank)
place1['final.rank'] = place1['region'].map(region_to_rank)
place2['final.rank'] = place2['region'].map(region_to_rank)

# Drop the temporary 'region' column if not needed
face1.drop(columns=['region'], inplace=True)
face2.drop(columns=['region'], inplace=True)
place1.drop(columns=['region'], inplace=True)
place2.drop(columns=['region'], inplace=True)
```

```python
# Extract the region name from the 'label' column
body1['region'] = body1['label'].str.split('_').str[-1]
body2['region'] = body2['label'].str.split('_').str[-1]
tool1['region'] = tool1['label'].str.split('_').str[-1]
tool2['region'] = tool2['label'].str.split('_').str[-1]

# Map the 'final.rank' from SA_rank to face1, face2, place1, and place2
region_to_rank = SA_rank.set_index('region')['final.rank']

body1['final.rank'] = body1['region'].map(region_to_rank)
body2['final.rank'] = body2['region'].map(region_to_rank)
tool1['final.rank'] = tool1['region'].map(region_to_rank)
tool2['final.rank'] = tool2['region'].map(region_to_rank)

# Drop the temporary 'region' column if not needed
body1.drop(columns=['region'], inplace=True)
body2.drop(columns=['region'], inplace=True)
tool1.drop(columns=['region'], inplace=True)
tool2.drop(columns=['region'], inplace=True)
```

OLS and spearman results for the split samples

```python
import statsmodels.api as sm
from scipy.stats import spearmanr
```

```python
# List of datasets to analyze
datasets = [face1, face2, place1, place2]
dataset_names = ['face_subsample1', 'face_subsample2', 'place_subsample1', 'place_subsample2']

# Iterate through each dataset
for dataset, name in zip(datasets, dataset_names):
    # Fit the model
    X = sm.add_constant(dataset['final.rank'])  # Add a constant for the intercept
    model = sm.OLS(dataset['Second_Mean'], X).fit()

    # Plot the original data
    plt.scatter(dataset['final.rank'], dataset['Second_Mean'], label='Original Data', color='blue')

    # Plot the fitted line
    plt.plot(dataset['final.rank'], model.predict(X), label='Fitted Line', color='red')

    plt.xlabel('Final Rank')
    plt.ylabel('Second Mean')
    plt.title(f'Linear Regression: Second Mean vs Final Rank ({dataset.iloc[0]["label"].split("_")[1]})')
    plt.suptitle(f'Dataset: {name}', fontsize=14, fontweight='bold')  # Add dataset name at the top
    plt.legend()
    plt.show()

    # Output the model summary
    print(model.summary())

    # Calculate Spearman's rho
    rho, p_value = spearmanr(dataset['Second_Mean'], dataset['final.rank'])
    print(f"Spearman's rho: {rho}, p-value: {p_value}")
```
